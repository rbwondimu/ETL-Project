```python
import pandas as pd
from sqlalchemy import create_engine 
```

```python
csv_file = "crime_2019.csv"
crime_2019_df = pd.read_csv(csv_file)
crime_2019_df.head()
```

```python
new_crime_2019_df = crime_2019_df [['REPORT_DAT', 'SHIFT', 'OFFENSE', 'BLOCK', 'WARD']].copy()
new_crime_2019_df.head()
```

```python
csv_file = "crime_2018.csv"
crime_2018_df = pd.read_csv(csv_file)
crime_2018_df.head()
```

```python
new_crime_2018_df = crime_2018_df [['REPORT_DAT', 'SHIFT', 'OFFENSE', 'BLOCK', 'WARD']].copy()
new_crime_2018_df.head()
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/crime_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

# Use pandas to load csv converted DataFrame into database

```python
new_crime_2019_df.to_sql(name ='new_crime_2019', con=engine, if_exists='append', index=False) 
```

# Use pandas to load csv2 converted DataFrame into database

```python
new_crime_2018_df.to_sql('new_crime_2018', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from new_crime_2019', con=engine).head()
```

```python
pd.read_sql_query('select * from new_crime_2018', con=engine).head()
```
